```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.model_selection import TimeSeriesSplit
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, Holt, ExponentialSmoothing
import warnings
warnings.filterwarnings('ignore')
```

Read data

```python
df = pd.read_csv("airline-passengers.csv")
df.head()
```

```python
df.info()
```

```python
df['Month'] = pd.to_datetime(df['Month']) # Change type of Month column to datetime
```

```python
fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(df['Month'], df['Passengers'])
```

creating the series

```python
s = pd.Series(index=df['Month'],
             data = df['Passengers'].values)
```

Decompose data to trend, season and residual

```python
result = seasonal_decompose(s, model='additive')
```

```python
fig, ax = plt.subplots(figsize=(7, 1))
ax.plot(result.trend.index, result.trend.values)
ax.set_title('Trend')

fig, ax = plt.subplots(figsize=(7, 1))
ax.plot(result.seasonal.index, result.seasonal.values, color='green')
ax.set_title('Seasonal')

fig, ax = plt.subplots(figsize=(7, 1))
ax.plot(result.resid.index, result.resid.values, color='red')
ax.set_title('Residual')
```

```python
fig, ax = plt.subplots()
ax.hist(result.resid.values)
```

Create a splitter for the series for crosvalidation

```python
tscv = TimeSeriesSplit(n_splits=5,test_size=12)
```

```python
#Visualise the splitting
for i, (train_index, test_index) in enumerate(tscv.split(s)):
    fig, ax = plt.subplots(figsize=(8, 2.7), layout='constrained')
    ax.plot(s.index, s.values, color='gray', label='Original data')
    train = s[train_index]
    test = s[test_index]
    ax.plot(train.index, train.values, color='blue', label='Train data')
    ax.plot(test.index, test.values, color='green', label='Test data')
    ax.legend()
```

Train on Holt model

```python
for i, (train_index, test_index) in enumerate(tscv.split(s)):
    train = s[train_index]
    test = s[test_index]
    holt = Holt(train)
    holt_fitted = holt.fit()
    preds = holt_fitted.forecast(len(test))
    
    print("r2 train", r2_score(train, holt_fitted.fittedvalues.values))
    print("r2 test",r2_score(test, preds))
    print("MAR train",mean_absolute_error(train, holt_fitted.fittedvalues.values))
    print("MAR test",mean_absolute_error(test, preds))
    print()
            
```

The Holt model doesn't predict well on test data. It's better to try another model.

Train on ExponentialSmoothing model

```python
for i, (train_index, test_index) in enumerate(tscv.split(s)):
    train = s[train_index]
    test = s[test_index]
    mod = ExponentialSmoothing(train,
                          trend = 'add',
                          seasonal='add',
                          seasonal_periods=12)
    mod_fit = mod.fit()
    preds = mod_fit.forecast(len(test))
    
    print("r2 train", r2_score(train, mod_fit.fittedvalues.values))
    print("r2 test",r2_score(test, preds))
    print("MAR train",mean_absolute_error(train, mod_fit.fittedvalues.values))
    print("MAR test",mean_absolute_error(test, preds))
    print()
```

```python
mean_metrics = {}
```

```python
trends = ['add', 'mul']
seasonals = ['add', 'mul']
mean_metrics = {}
for i in trends:
    for j in seasonals:
        MAR_train = {}
        MAR_test = {}
        R2_train = {}
        R2_test = {}
        for k, (train_index, test_index) in enumerate(tscv.split(s)):
            train = s[train_index]
            test = s[test_index]
            mod = ExponentialSmoothing(train,
                                      trend = i,
                                      seasonal = j,
                                      seasonal_periods=12)
            mod_fit = mod.fit()
            preds = mod_fit.forecast(len(test))
            R2_train[k] = r2_score(train, mod_fit.fittedvalues.values)
            R2_test[k] = r2_score(test, preds)
            MAR_train[k] = mean_absolute_error(train, mod_fit.fittedvalues.values)
            MAR_test[k] = mean_absolute_error(test, preds)
        #print("Mean values for trend {} and seasonal {}:\n r2_train: {}, \n r2_train: {}, \n MAR_train: {}, \n MAR_test: {}".format(i, j, sum(R2_train.values())/len(R2_train), sum(R2_test.values())/len(R2_test), sum(MAR_train.values())/len(MAR_train), sum(MAR_test.values())/len(MAR_test)))
        key = i +''+ j
        mean_metrics[key] = {'r2_train': sum(R2_train.values())/len(R2_train), 'r2_test': sum(R2_test.values())/len(R2_test), 'MAR_train': sum(MAR_train.values())/len(MAR_train), 'MAR_test': sum(MAR_test.values())/len(MAR_test)}
        
```
